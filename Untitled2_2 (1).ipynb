Доказать, что функция f(x) непрерывна в точке x_0 (найти δ(ε) ).

f(x) = 3x^2-3  x0 = 4

Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
```

Рассмотрим функцию

```python
def f_x(x):
    y = 3 * x ** 2 - 3
    return y
```

Обозначим переменные и найдем delta и eps.

```python
delta = 10
dx = 1e-5
eps = 0
x_0 = 4
y_0 = f_x(x_0)

while delta > dx:
    eps -= 1
    delta = f_x(x_0 + 10 ** eps) - y_0
liml = round(f_x(x_0 - 10 ** eps))
while delta > dx:
    eps -= 1
    delta = f_x(x_0 - 10 ** eps) - y_0
limr = round(f_x(x_0 + 10 ** eps))
if (limr == liml == y_0): 
    print(f'Функция непрерывна в точке "{x_0}"')
else: 
    print(liml, y_0, limr)
```

построим график

```python
x = np.linspace(x_0 - 1, x_0 + 1, 100)
y = f_x(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.plot(x, y, color='black', linewidth=0.5)
ax.scatter(x_0, f_x(x_0), color='red', marker='*')
plt.show()
```
