Задание 1 номер 1
Доказать, что lim a_n = a при n → ∞
a_n = (7n + 4)/(2n + 1), a = 3/2

```
import matplotlib.pyplot as plt
```

```
flag = True
lim_a = 7 / 2
width = 100
x, y = [], []
epsilon = 1e-7
n = 1
while flag == True:
     a_n = (7 * n + 4) / (2 * n + 1)
     delta = abs(lim_a - a_n)
     n += 1
     if len(x) < width:
         x.append(n)
         y.append(a_n)
     if delta < epsilon: 
         flag = False
print(f"Предел функции a_n = {lim_a}. ")
```
