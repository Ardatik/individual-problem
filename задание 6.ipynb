Произведем импорт необходимых библиотек.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Рассмотрим функцию.

```python
def f(x):
  return (2*((x-2) ** 2) * (8-x))**1/3
```

Создадим переменные x и y.

```python
a = 0
b = 6
x = np.linspace(a, b, 30)
y = f(x)
```

Построим график функции.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.plot(x, y, color = 'black')
plt.show()
```

Найдем наибольшее и наименьшее значение функции на отрезке [0;6].

```python
a = 0
b = 6
d_x = 0.1
x_1 = (a + b - d_x) / 2
x_2 = (a + b + d_x) / 2
eps = 0.05

while (b - a) / 2 > eps:
  if f(x_1) <= f(x_2):
    a = x_1
    x_1 = (a + b - d_x) / 2
    x_2 = (a + b + d_x) / 2
  else:
    b = x_2
    x_1 = (a + b - d_x) / 2
    x_2 = (a + b + d_x) / 2
max_znach = (a + b) / 2
print(f'Наибольшее значение функции - "{max_znach}"')

a = 0
b = 6
d_x = 0.1
x_1 = (a + b - d_x) / 2
x_2 = (a + b + d_x) / 2
eps = 0.06

while (b - a) / 2 > eps:
  if f(x_1) <= f(x_2):
    b = x_2
    x_1 = (a + b - d_x) / 2
    x_2 = (a + b + d_x) / 2
  else:
    a = x_1
    x_1 = (a + b - d_x) / 2
    x_2 = (a + b + d_x) / 2
min_znach = (a + b) / 2
print(f'Наименьшее - "{min_znach}"')
```

Добавим точки наибольшего и наименьшего значения функции на график.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
x_min = np.linspace(min_znach, min_znach)
y_min = np.interp(x_min, x, y)
x_max = np.linspace(max_znach, max_znach)
y_max = np.interp(x_max, x, y)
ax.plot(x, y, color = 'black')
ax.plot(x_min, y_min, '*', color = 'red')
ax.plot(x_max, y_max, '*', color = 'blue')
plt.show()
```
