Доказать (найти δ(ε) ).

lim (3x^2 + 5x - 2) / (x + 2) = -7 при x → -2

Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
```

Рассмотрим функцию

```python
def f_x(x):
    y = (3*x**2 + 5*x - 2) / (x + 2)
    return y
```

Обозначим переменные и найдем delta и eps.

```python
delta = 10
dx = 1e-5
eps = 0
x_0 = -3

while delta > dx:
    eps -= 1
    delta = f_x(x_0 + 10 ** eps) - f_x(x_0 - 10 ** eps)
print(f'Значение delta - "{delta}"\nЗначение eps, при котором delta принимает вид - "{eps}"')
```

построим график

```python
x = np.linspace(x_0 - 0.5, x_0 + 0.5, 100)
y = f_x(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.plot(x, y, color='black', linewidth=0.5)
ax.scatter(x_0, f_x(x_0 + 10 ** eps), color='red', marker='*')
plt.show()
```
