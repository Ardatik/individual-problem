Вычислить приближенно при помощи дифференциала.

y = (x + (5-x**2)**1/2)/2 , x = 0.98

Рассмотрим функцию.

```python
def f_x(x: float):
    return (x + (5-x**2)**1/2)/2
```

Найдем производную

```python
x = 0.98
x_0 = 1
delta_x = x - x_0
y_0 = f_x(x_0)
dx = 1e-6
y = f_x(x)

diff_y = (f_x(x_0 + dx/2) - f_x(x_0 - dx/2)) / dx
print(f'Значение производной в точке x_0 - "{diff_y}"')
```

подсчитаем примерное значение, а также сравним с прямым вычислением

```python
y_x = y + diff_y * delta_x
print(f'Примерное значение с помощью дифференциала в точке x - "{y_x}"')
print(f'Разность, получаемая при сравнении прямого вычисления и приближенного - "{f_x(x) - y_x}"')
print(f_x(x))
```
