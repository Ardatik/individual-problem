Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

Создание функции

```python
def fx(x: float):
    return (2 * ((x-2)**2) * (8-x) )**1/3
```

Строим функцию по точкам

```python
a = 0 
b = 6
x = [i / 10 for i in range(10 * a, 10 * b)]
y = [fx(i) for i in x]
```

Ищем точки минимума и максимума

```python
dx = 0.1
x1 = (a + b - dx) / 2
x2 = (a + b + dx) / 2
eps = 0.07
while (b - a) / 2 > eps:
   if fx(x1) < fx(x2):
      b = x2
      x1 = (a + b - dx) / 2
      x2 = (a + b + dx) / 2
   else:
    a = x1
    x1 = (a + b - dx) / 2
    x2 = (a + b + dx) / 2

print(f'точка минимума » ({round((a + b)/2, 1)}, {fx(round((a + b)/2, 1))})')

a = 1
b = 2
dx = 0.1
x1 = (a + b - dx) / 2
x2 = (a + b + dx) / 2
eps = 0.07
while (b - a) / 2 > eps:
  if fx(x1) < fx(x2):
    a = x1
    x1 = (a + b - dx) / 2
    x2 = (a + b + dx) / 2
  else:
    b = x2
    x1 = (a + b - dx) / 2
    x2 = (a + b + dx) / 2
```

```python
fig = plt.figure(10)
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.plot(x, y, linewidth=1.0, color = 'black')
plt.scatter(round((a + b)/2, 1), round(fx((a + b)/2), 1), color = 'red')
plt.scatter(round((a - b)/2, 1), round(fx((a - b)/2), 1), color = 'blue')
print(f'точка максимума » ({round((a + b)/2, 1)}, {fx(round((a + b)/2, 1))})')


plt.show()
```
