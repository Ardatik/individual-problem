Доказать, что lim a_n = a при n → ∞

a_n = (7n + 4)/(2n + 1), a = 7/2

Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
```

обозначим переменные и найдём предел функции

```python
lim_a = 7 / 2
eps = 1e-3
N_eps = 1
width = 100
x, y = [], []
flag = True

while flag == True:
     a_n = (7 * N_eps + 4) / (2 * N_eps + 1)
     delta = abs(lim_a - a_n)
     N_eps += 1
     if len(x) < width:
         x.append(N_eps)
         y.append(a_n)
     if delta < eps: 
         flag = False
print(f"Предел функции к которому стремится a_n = {a_n} ,дельта, при котором функция стремится к пределу = {delta}, N_eps, при которой дельта имеет вид {N_eps}")
```

построим график функции

```python
fig = plt.figure(10)
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.scatter(x, y, color='blue', marker='.', linewidth=0.5)
ax.plot(np.linspace(0, width + 1), np.linspace(7/2, 7/2), linewidth=1.5, color='red')

plt.show()
```

Таким образом мы видим, что при наращивании x, y стремится к точке 3/2, но не достигает её
