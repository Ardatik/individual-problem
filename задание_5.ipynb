Импортируем библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

определим функцию

```python
def fx(x: float):
    return x**2*(x-2)**2
```

Считаем производные 1 и 2 порядка

```python
dx = 10**(-5)
def dir(x0):
    return (fx(x0 + dx) - fx(x0)) / dx


def dir_2(x):
    return (dir(x + dx) - dir(x)) / dx

print(dir_2(1))
```

Составляем график по точкам и ищем точки, в которых первая производная равна 0

```python
a=[]
p=np.linspace(-1.5,1.5,1050)
y1=[fx(i) for i in p]

for i in p:
  if round(dir(i),1)==0:
    a.append(i)
x0=round(a[-1],2)
print(a,x0)
```

Записываем формулу Тейлора

```python
y2=[fx(x0)+dir(x0)*(x-x0)+dir_2(x0)*((x-x0)**2)/2 for x in p]
```

Строим график по точкам и по формуле Тейлора

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.grid(True)

ax.plot(p,y1,color="black")
ax.plot(p,y2,color="red")
plt.scatter(x0,fx(x0),color="black")
plt.show()
```
